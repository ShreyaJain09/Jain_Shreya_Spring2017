```python
from nltk.corpus import webtext
fileid= webtext.fileids()
#for fileid in webtext.fileids():
print(fileid)
```

```python
jack=0
will=0
for pick in webtext.words('pirates.txt'):
    if pick =='JACK':
        jack =jack + 1
    elif pick =='WILL':
        will =will + 1
print("Jack Sparrow's lines are:", jack)
print("Will Turner's lines are", will)
if jack>will:
        print('Leading actor of Pirrates of Carribean is Jack Sparrow')
else:
        print('Leading actor of Pirates of Carribean is Will Turner')
```
