```python
#Calculte the frequency of all the words including stopwords and punctuations
from collections import Counter
counts = Counter(pers)
```

```python
print(counts)
```

```python
#Top 10 most frequent words and their frequency
counts.most_common(10)
```

```python
freqs = {}
for word in pers:
    freqs[word] = freqs.get(word, 0) + 1 
```
