Reading data from NYTAPI - MostPopular API

```python
import requests # to request data from api
import time #to set time to sleep api calls
#creating url to fetch data from MostPopular API for all three categories (MostViewed, MostShared, MostEmailed)
#MostEmailed URL
url= 'https://api.nytimes.com/svc/mostpopular/v2/mostemailed/all-sections/30.json?api-key=8171f9b6eba24d638bd7986a927e3a35'
#MostViewed URL
viewed_url= 'https://api.nytimes.com/svc/mostpopular/v2/mostviewed/all-sections/30.json?api-key=8171f9b6eba24d638bd7986a927e3a35'
#MostShared URL
shared_url= 'https://api.nytimes.com/svc/mostpopular/v2/mostshared/all-sections/30.json?api-key=8171f9b6eba24d638bd7986a927e3a35'
```

```python
popular = [] #List for collecting all data
def myFunction(json_data_bsh):    # This function extracts required data from the tags in API
    for i in json_data_bsh['results']:
        dic = {}
        dic['title'] = i['title']
        dic['section'] = i['section']
        dic['published_date'] = i['published_date']
        dic['byline'] = i['byline']
        popular.append(dic)
```

APIs populated only 20 results. So we made multiple calls by setting offset value(multiple of 20) in a loop.

```python
#fetching results for MostEmailed 
pages=40
for i in range(0, pages):
    offset = i*20
    myurl = url+ '&' + 'offset=' + str(offset)
    r = requests.get(myurl)
    json_data = r.json()
    myFunction(json_data)
    time.sleep(0.5)
```

```python
#fetching results for MostShared 
pages=71
for i in range(0, pages):
    offset = i*20
    myurl = shared_url+ '&' + 'offset=' + str(offset)
    r = requests.get(myurl)
    json_data = r.json()
    myFunction(json_data)
    time.sleep(0.5)
```

```python
#fetching results for MostViewed
pages=93
for i in range(0, pages):
    offset = i*20
    myurl = viewed_url+ '&' + 'offset=' + str(offset)
    r = requests.get(myurl)
    json_data = r.json()
    myFunction(json_data)
    time.sleep(0.5)
```

Saving data to local disk in CSV format

```python
import unicodecsv as csv # to read/write csv data
keys = popular[0].keys() # Getting the keys from dictionary

#writing data in csv format at local disk
with open('popular_3.csv', 'wb') as output_file:
    dict_writer = csv.DictWriter(output_file, keys)
    dict_writer.writeheader()
    dict_writer.writerows(popular)
```
