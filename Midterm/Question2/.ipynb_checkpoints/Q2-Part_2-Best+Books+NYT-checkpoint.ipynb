Reading data from NYT API - BestSeller Books

```python
import requests # to request data from APIs
nyt_url= 'https://api.nytimes.com/svc/books/v3/lists/best-sellers/history.json?api-key=4e1e0fdeb9c647fe9532b22a1fe26075' 
```

```python
best_books = [] # defining list for collecting data
```

```python
def myFunction(json_data_bsh): #this fuction collects data from json and writes it to a list
    for i in json_data_bsh['results']:
        dic = {}
        dic['title'] = i['title']
        dic['description'] = i['description']
        dic['contributor'] = i['contributor']
        dic['author'] = i['author']
        dic['contributor_note'] = i['contributor_note']
        dic['price'] = i['price']
        dic['age_group'] = i['age_group']
        dic['publisher'] = i['publisher']
        for x in i['isbns']:
            dic['isbn10'] = x['isbn10']
            dic['isbn13'] = x['isbn13']
        for x in i['ranks_history']:
            dic['primary_isbn10'] = x['primary_isbn10']
            dic['primary_isbn13'] = x['primary_isbn13']
            dic['rank'] = x['rank']
            dic['list_name'] = x['list_name']
            dic['display_name'] = x['display_name']
            dic['published_date'] = x['published_date']
            dic['bestsellers_date'] = x['bestsellers_date']
            dic['weeks_on_list'] = x['weeks_on_list']
            dic['ranks_last_week'] = x['ranks_last_week']
            dic['asterisk'] = x['asterisk']
            dic['dagger'] = x['dagger']
        for x in i['reviews']:
            dic['book_review_link'] = x['book_review_link']
            dic['first_chapter_link'] = x['first_chapter_link']
            dic['sunday_review_link'] = x['sunday_review_link']
            dic['article_chapter_link'] = x['article_chapter_link']
        best_books.append(dic) # appending individual record here
```
