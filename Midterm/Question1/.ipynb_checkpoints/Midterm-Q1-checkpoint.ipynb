# Analysis 1: 
To check the top 10 people who the CEO Kenneth Lay sent out emails to and who did he receive the emails most from.
Also gives the count of how many emails to which individual and how many from whom.

```python
#Code is same as above except that some of the fields are empty 
#(due to data corruption, old Mime formats that can’t be parsed)
#adding all the emails separately when there are multiple id's in to
import os
from collections import Counter

from email.parser import Parser
rootdir = "~/midterm/data/enron/maildir/lay-k"


def email_analyse(inputfile, to_email_list, from_email_list, email_body):
    with open(inputfile, "r",encoding='utf-8', errors='ignore') as f:
        data = f.read()

    email = Parser().parsestr(data)
    #remove all the newline characters (\n), all the tabs (\t) and empty spaces and to make sure email is parsed correctly 
    if email['to']:
        email_to = email['to']
        email_to = email_to.replace("\n", "")
        email_to = email_to.replace("\t", "")
        email_to = email_to.replace(" ", "")

        email_to = email_to.split(",")

        #loop over the list and append it to the to_email_list
        for email_to_1 in email_to:
            to_email_list.append(email_to_1)

    from_email_list.append(email['from'])

    email_body.append(email.get_payload())


to_email_list = []
from_email_list = []
email_body = []

for directory, subdirectory, filenames in  os.walk(rootdir):
    for filename in filenames:
        email_analyse(os.path.join(directory, filename), to_email_list, from_email_list, email_body )

#to find out who the top 10 most popular addresses use counter
print("\nTo email adresses: \n")
print(Counter(to_email_list).most_common(10))

print("\nFrom email adresses: \n")
print(Counter(from_email_list).most_common(10))


with open("to_email_list.txt", "w") as f:
    for to_email in to_email_list:
        if to_email:
            f.write(to_email)
            f.write("\n")
with open("from_email_list.txt", "w") as f:
    for from_email in from_email_list:
        if from_email:
            f.write(from_email)
            f.write("\n")        
with open("email_body.txt", "w") as f:
    for email_bod in email_body:
        if email_bod:
            f.write(email_bod)
            f.write("\n")   
            
#renamed the file email_body.txt to ken_lay_emails.txt
```

# Analysis 3


```python
#remove the punctuations, stopwords and normalize the corpus, this is to form the dictiornary to do LDA analysis.
from nltk.corpus import stopwords 
from nltk.stem.wordnet import WordNetLemmatizer
import string

with open("~/midterm/Question1/Output/ken_lay_emails.txt", "r",encoding="utf-8", errors="ignore") as f:
    data = f.read()

stop = set(stopwords.words('english'))
exclude = set(string.punctuation) 
lemma = WordNetLemmatizer()
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
doc_clean = [clean(doc).split() for doc in data] 
```

```python
# Importing Gensim
import gensim
from gensim import corpora
```

```python
## Creating the term dictionary of our courpus, where every unique term is assigned an index.
dictionary = corpora.Dictionary(doc_clean)
```

```python
# Converting list of documents (corpus) into Document Term Matrix using dictionary prepared above.
doc_term_matrix = [dictionary.doc2bow(doc) for doc in doc_clean]
```

```python
# Creating the object for LDA model using gensim library
Lda = gensim.models.ldamodel.LdaModel
```

```python
import lda
```

```python
# Running and Trainign LDA model on the document term matrix.
ldamodel = Lda(doc_term_matrix,num_topics = 50, update_every = 1, chunksize = 100000, passes = 350)
```
