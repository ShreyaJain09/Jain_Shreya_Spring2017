```python
posts = nltk.corpus.nps_chat.xml_posts()
type(posts)
counter = 0
for all in posts:                           
    shreya = all.get("user")                   
    if shreya.endswith("User121"):
        counter +=1

print('The total number of posts by user121 is ', counter)
```

```python
import re
from os.path import basename
from os.path import splitext
```

```python
s_20 = []
s_30 = []
s_40 = []
s_adults = []
s_teens = []

for file in nltk.corpus.nps_chat.fileids():
    X = nltk.corpus.nps_chat.raw(file)
    base = basename(file)
    base = splitext(base)[0][6:9]
    if base in '20s':
       for line in X.split('\n'):
           if line.strip().startswith('<Post '):
             s_20.append(re.findall(r'User([^(]*)\">', line))
             
    if base in '30s':
       for line in X.split('\n'):
           if line.strip().startswith('<Post '):
             s_30.append(re.findall(r'User([^(]*)\">', line))
              
    if base in '40s':
       for line in X.split('\n'):
           if line.strip().startswith('<Post '):
             s_40.append(re.findall(r'User([^(]*)\">', line))
              
    if base in 'adults':
        for line in X.split('\n'):
           if line.strip().startswith('<Post '):
             s_adults.append(re.findall(r'User([^(]*)\">', line))
             
    if base in 'teens':    
        for line in X.split('\n'):
           if line.strip().startswith('<Post '):
             s_teens.append(re.findall(r'User([^(]*)\">', line))
             
s_20=sum(s_20,[])
s_30=sum(s_30,[])
s_40=sum(s_40,[])
s_adults=sum(s_adults,[])
s_teens=sum(s_teens,[])        

print("*****UNIQUE USERS PER GROUP*****")
print("\nNumber of Unique Users in age group 20s   :",len(set(s_20)))
print("\nNumber of Unique Users in age group 30s   :",len(set(s_30)))
print("\nNumber of Unique Users in age group 40s   :",len(set(s_40)))
print("\nNumber of Unique Users in age group Adults:",len(set(s_adults)))
print("\nNumber of Unique Users in age group Teens :",len(set(s_teens)))
```
